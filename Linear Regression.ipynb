<h1>Linear Regression models on the housing dataset</h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn import metrics
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#Reading the csv and making a dataframe from the data
housing = pd.read_csv(r"C:\Users\myste\Documents\MSCS\Fall 2020\ML\HousingDataset.csv")
housing.head()
```

<h2> Checking the corelation</h2>

```python
#The best graph to check the correlation is by plotting a heatmap for the columns
fig = plt.gcf() 
figsize = fig.get_size_inches()
fig.set_size_inches(figsize * 2.5)
sns.heatmap(housing.corr(),annot=True)
plt.show()
```

<h2>As we can see from the above graph Tax has very strong correlation with Rad, Nox and Indus. We can use these to predict Tax.</h2>

<h2>Separating the data into features and target variable and test and train data</h2>

```python
X=housing[['RAD','NOX','INDUS']]
y=housing['TAX']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

<h2>Linear Regression:</h2>

```python
linear=LinearRegression()
linear.fit(X_train,y_train)
predicted_linear = linear.predict(X_test) 
print("Accuracy:",linear.score(X_test, y_test))
#Plots for predictions
figure, ax = plt.subplots()
ax.scatter(y_test, predicted_linear)
ax.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
ax.set_title("Actual Truth vs Predicted")
plt.show()
```

<h2>Ridge Regression:</h2>

```python
ridge=Ridge()
ridge.fit(X_train,y_train)
predicted_ridge = ridge.predict(X_test)
print("Accuracy:",ridge.score(X_test, y_test))
#Plots for predictions
figure, ax = plt.subplots()
ax.scatter(y_test, predicted_ridge)
ax.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
ax.set_title("Actual Truth vs Predicted")
plt.show()
```

<h2>Lasso Regression:</h2>

```python
lasso=Lasso()
lasso.fit(X_train,y_train)
predicted_lasso = lasso.predict(X_test)
print("Accuracy:",lasso.score(X_test, y_test))
#Plots for predictions
figure, ax = plt.subplots()
ax.scatter(y_test, predicted_lasso)
ax.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
ax.set_title("Actual Truth vs Predicted")
plt.show()
```

<h2>Elastic Net:</h2>

```python
en=ElasticNet()
en.fit(X_train,y_train)
predicted_en = en.predict(X_test)
print("Accuracy:",en.score(X_test, y_test))
#Plots for predictions
figure, ax = plt.subplots()
ax.scatter(y_test, predicted_en)
ax.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
ax.set_title("Actual Truth vs Predicted")
plt.show()
```

<h1>Results and Comparision:</h1>

<h3>Random Forest has the best accuracy followed by Decision Tree. Support Vector has the worst accuracy.<br><br>
As we can see from the graphs the models where values don't lie on the predicted value line which have more deviations give us less accuracy.<br><br> When we predict TAX it had higher accuracies because the features we selected are more correlated. NOX has less correlation to features used to predict it and hence it has slightly lower accuracies.<br><br>
The models and their explainations are below:<br><br>
    1.Linear Regression works well if there is a linear relationship between variables. It fits a linear model with weight coefficients to minimize the residual sum of squares between the observed targets in the dataset, and the targets predicted by the linear approximation. <br><br>
    2. Ridge Regression is a regularization method and uses l2 regularization. It imposes a penalty on the size of the coefficients. The ridge coefficients minimize a penalized residual sum of squares. So, the coefficients become more robust to collinearity.<br><br>
    3. Lasso Regression minimizes the coefficients to zero. This is a regularization method and uses l1 regularization. If group of predictors are highly correlated, it picks only one and minimizes the others to zero thus effectively reducing the number of features upon which the given solution is dependent.<br><br>
    4.Elastic Net encourages group effect in case of highly correlated variables and can be prone to double shrinkage. It is trained with both l1 and l2-norm regularization of the coefficients. This combination allows for learning a sparse model where few of the weights are non-zero like Lasso, while still maintaining the regularization properties of Ridge.<br><br>
    5.Decision Tree algorithm is a supervised learning algorithms. But, it can be used to solve regression and classification problems as well. Decision tress understand the data and make some interpretations. Since decision tree makes a multi level tre for classification and predictions its very accurate.<br><br>
    6. Since random forest regression works by combining various decision trees to form an ensemble random forest, it has a better performance than an individual decision tree due to randomness, which helps to decrease variance. Hence the accuracy by this method is the best.<br><br>
    7. SVR only weighs instances (support vectors), not features. Due to this the application become really noisy and prone to random systems. This decreases the accuracy for this method. To increase accuracy we can change the number of features as it highly depends on the number of features.</h3>
